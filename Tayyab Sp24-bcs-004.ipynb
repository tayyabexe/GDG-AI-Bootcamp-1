Step 2: Handling Missing Values

Check for missing values in each column

```python
print(df.isnull().sum())
```

Filling director, country, cast with "Unknown" as the data has a huge percent of missing values. < 0.5%

```python
df['director'] = df['director'].fillna('Unknown')
df['country'] = df['country'].fillna('Unknown')
df['cast'] = df['cast'].fillna('Unknown')
```

Dropping the date_added, rating and duration as their contribution is less

```python
df.dropna(subset=['date_added', 'rating','duration'],inplace =True)
```

```python
print(df.shape)
print(df.isnull().sum())
```

Step 3: Convert dates and extract year

```python
# Strip spaces and THEN convert to make sure there are no leading/trailing spaces in the date strings, causing coerce errors
df['date_added'] = df['date_added'].str.strip()
df['date_added']=pd.to_datetime(df['date_added'], errors='coerce')
df
```

```python
# It tells how many values were coerced to NaT during the conversion
coerce_count = df['date_added'].isna().sum()
print(f"Number of coerced (NaT) values: {coerce_count}")
```

```python
#Year extraction from the date_added column
df['year_added']= df['date_added'].dt.year
```

```python
# just to verify no nulls remain in critical columns
print(df.isnull().sum())
```

Extract release year (ensure numeric)


```python
df['release_year'] = pd.to_numeric(df['release_year'], errors='coerce')
```

```python
# It tells how many values were coerced to NaT during the conversion
coerce_countForReleaseYear = df['release_year'].isna().sum()
print(f"Number of coerced (NaT) values: {coerce_countForReleaseYear}")
```

Step 4: Process multi-value columns

```python
# splitting countries into list
df['country_list'] = df['country'].apply(lambda x: [i.strip() for i in x.split(',')] if pd.notnull(x) else [])

# Splitting the genres
df['genre_list'] = df['listed_in'].apply(lambda x: [i.strip() for i in x.split(',')] if pd.notnull(x) else [])
```

```python
df.to_csv("cleaned_data.csv", index=False)
```

Question 2: Top 10 countries prodcuing most netflix content

```python
# now here we have to explode the country_list field and we will add it 
# into another table for better analysis
df_countries = df[['show_id','country_list','type']].copy()
df_countries = df_countries.explode('country_list')
df_countries['country_list']=df_countries['country_list'].str.strip()
#Currently we have 'Unknown' as country for missing values, we will remove those rows
df_countries = df_countries[df_countries['country_list'] != 'Unknown']
# we will use index so to get the names of the countries , not their counts
top_10_names = df_countries['country_list'].value_counts().head(10).index 

df_top_10= df_countries[df_countries['country_list'].isin(top_10_names)] 
# we have got the only top 10 countries with most content on netflix , along with their type of content
df_top_10
```

```python
plt.figure(figsize=(12,8))
sns.countplot(data=df_top_10, y='country_list', hue='type', order=top_10_names, palette='Set2')
plt.title('Top 10 Countries by Number of Titles on Netflix')
plt.xlabel('Number of Titles')
plt.ylabel('Countries')
plt.legend(title='Type of Content')
plt.show()
```

Analysis insights:
1. US is the bigger content supplier, and produced more movies than the TV Shows that are listed on netflix
2. Simlarly, India has higher ratio of Movies than TV Shows. While Japan and South Korea are focused on producing TV Shows than movies.

Question 7: Has Netflix expanded into new genres over time, or is it repeatedly investing in the same genres?
Support your answer using year-wise genre diversity analysis.


```python
df_genre_diversity = df.explode('genre_list')
df_genre_diversity['genre_list'] = df_genre_diversity['genre_list'].str.strip()
diversity_trend = df_genre_diversity.groupby('year_added')['genre_list'].nunique().reset_index() #this tells how many unique genres were added each year
top_genres_yearly = df_genre_diversity.groupby(['year_added', 'genre_list']).size().unstack(fill_value=0) # this tells frequency of each genre per year


plt.figure(figsize=(12, 6))
sns.lineplot(data=diversity_trend, x='year_added', y='genre_list', marker='o', color='#E50914', linewidth=2)
plt.title('Number of Unique Genres Added Per Year')
plt.xlabel('Year Added')
plt.ylabel('Count of Unique Genres')
plt.grid(True, linestyle='--', alpha=0.6)
plt.show()

top_10_genres_list = df_genre_diversity['genre_list'].value_counts().head(10).index
subset_heatmap = top_genres_yearly[top_10_genres_list].tail(10) # last 10 years

plt.figure(figsize=(14, 8))
sns.heatmap(subset_heatmap.T, annot=True, fmt='d', cmap='YlGnBu')

plt.title('Heatmap: Investment in Top Genres Over Time')
plt.xlabel('Year Added')
plt.ylabel('Genre')
plt.show()
```

Answer: Netflix invests in both the new genres over time and the same genres. It used a "Big Expansion" strategy followed by a "Safe Investment" strategy.

1. The Big Expansion (2014 – 2016)

    Looking at the Red Line Chart (Number of Unique Genres Added Per Year):
    What is happening? Between 2014 and 2016, the line goes up very sharply. It jumps from about 15 unique genres to 42 unique genres in just two years.
    
    This proves that Netflix actively expanded its variety. It didn't just add more of the same; it started adding completely new categories (like Anime, specialized Documentaries, and International shows) to attract different types of people around the world.

2. Strategic Repetition (2017 – 2021)

    Looking at the Heatmap (Investment in Top Genres Over Time):

    Genres like "International Movies," "Dramas," and "Comedies" consistently have the highest numbers every single year. For example, "International Movies" jumped from 82 titles in 2016 to over 600 titles in later years.

    This shows repeated investment. Once Netflix found out that people love Dramas and International content, they stopped looking for "new" genres and started "repeating" their success by pouring more money into these specific categories.


Question 8: Is Netflix overly dependent on a small number of countries for its content supply?
Analyze this using content share ratios rather than raw counts.


```python
total_productions = len(df_countries)
#getting the percentage shae for each country
country_share_ratios = (df_countries['country_list'].value_counts() / total_productions) * 100
top_10_ratios = country_share_ratios.head(10)
others_ratio = country_share_ratios.iloc[10:].sum()
plot_shares = top_10_ratios.copy()
plot_shares['Others'] = others_ratio
```

```python
plt.figure(figsize=(10, 10))
colors = plt.cm.Paired(range(len(plot_shares)))
plt.pie(plot_shares, labels=plot_shares.index, autopct='%1.1f%%', 
        startangle=140, colors=colors, pctdistance=0.85, explode=[0.05]*11)
centre_circle = plt.Circle((0,0), 0.70, fc='white')
fig = plt.gcf()
fig.gca().add_artist(centre_circle)
plt.title('Q8: Content Share Ratio by Country',)
plt.tight_layout()
plt.show()
```

Answer: Yes, the Netflix relies heavily on a handful of nations, particularly the United States, to fill its library.

1. The United States alone supplies 36.8% of all content. This means more than one out of every three titles you see on Netflix comes from just one country.

2. Just three countries—the United States (36.8%), India (10.5%), and the United Kingdom (8.0%)together provide over 55% of Netflix's entire catalog. This shows that the majority of the platform's supply is concentrated in a very small part of the world.

3. The top 10 countries combined produce roughly 75% of all content. All other countries in the world combined (the "Others" slice) only account for about 24.6%. This proves Netflix is highly dependent on a small elite group of content-producing hubs.

Question 9: Has there been a noticeable shift toward more mature content ratings over the years?
Compare trends for Movies and TV Shows separately.


```python
target_ratings = ['TV-MA', 'TV-14', 'TV-PG', 'R']
df_filtered = df[df['rating'].isin(target_ratings)]
#Grouping by year, type, and rating
rating_trend = df_filtered.groupby(['year_added', 'type', 'rating']).size().reset_index(name='count')

fig, ax = plt.subplots(1, 2, figsize=(16, 6))
# plot for movies
sns.lineplot(data=rating_trend[rating_trend['type'] == 'Movie'], 
             x='year_added', y='count', hue='rating', ax=ax[0], marker='o')
ax[0].set_title('Movie Rating Trends Over Time')
ax[0].grid(True, linestyle='--', alpha=0.5)

# plot for TV shows
sns.lineplot(data=rating_trend[rating_trend['type'] == 'TV Show'], 
             x='year_added', y='count', hue='rating', ax=ax[1], marker='o')
ax[1].set_title('TV Show Rating Trends Over Time')
ax[1].grid(True, linestyle='--', alpha=0.5)

plt.tight_layout()
plt.show()
```

Analysis: 
1. Since 2016, there has been an explosive shift toward mature ratings. TV-MA has transitioned from being a minor category to the most dominant rating added to the platform every year.

2. The data shows that Netflix is aggressively targeting adult audiences. While teen-rated content (TV-14) is also growing, it is being significantly outpaced by mature-rated content (TV-MA and R).


Question 10: Based on your EDA findings, identify one potential strategic risk in Netflix’s content strategy and justify it using data-driven evidence.



Answer:

    A significant strategic risk for Netflix is its heavy over-reliance on mature-rated movies compared to long-term, family-friendly TV content.

    While Netflix has built a massive library, the current distribution creates a risk of higher subscriber churn (cancellation) because the content mix is skewed toward "one-and-done" viewing experiences.

1. Movies currently dominate the platform, making up 69.7% of the total catalog, while TV Shows account for only 30.3%. From a strategy perspective, TV shows are generally better for retention because they keep users subscribed over several weeks or months, whereas movies are consumed in a single sitting.

2. Netflix’s content is heavily skewed toward older audiences. The most common rating is TV-MA (3,205 titles), followed by TV-14 (2,157 titles). In contrast, family-friendly ratings like TV-G and G account for a tiny fraction of the library (220 and 41 titles respectively). This limits the platform's "stickiness" for households with children, who may prioritize competitors like Disney+.

This report summarizes our journey of cleaning and analyzing the Netflix dataset, telling the story from the raw data to the final strategic insights we discovered.

1. The Dataset

    The dataset was a collection of information about movies and TV shows available on Netflix. It included details like the title, release year, country of origin, and genre. Our goal was to understand how Netflix’s content library has changed over time and identify risks in their current strategy.

2. Problems in the Data
    
     When we first looked at the data, it wasn't ready for analysis. We found several issues:
    - Missing Values: Many columns had empty (null) cells, especially the director, cast, and country columns.
    - Incorrect Data Types: The date_added column was stored as text instead of a date format, making it impossible to track trends over time.
    - Irrelevant Data: Some columns, like director, had so many missing values that they were not useful for a broad analysis.

3. How the Data Was Cleaned

    To make the data reliable, we followed these cleaning steps:
    - Handling Nulls: We dropped the director column because it was mostly empty. For cast and country, we filled missing values with "Unknown" so we wouldn't lose those rows.
    - Date added, duration and rating field's missing values were dropped.
    - We converted the date_added column into a proper datetime format. This allowed us to extract the "Year Added" for our trend charts.
    - Structural Check: We ensured there were no duplicate rows that could skew our results.

4. Columns Worked On

    We worked with a total of 12 columns:
    - 1. show_id | 2. type | 3. title | 4. director | 5. cast | 6. country | 7. date_added | 8. release_year | 9. rating | 10. duration | 11. listed_in (Genres) | 12. description.

5. How the Analysis Was Done
    
    We used Exploratory Data Analysis (EDA). We didn't just look at totals; we looked at how variables related to each other. For example:

    - We compared the volume of Movies vs. TV Shows.
    - We tracked how many new titles were added each year.
    - We looked at which genres were getting the most investment.

6. Charts and Plots Made

    To visualize our findings, we created several specific plots:

    - Pie Chart: To show the distribution of Movies vs. TV Shows.
    - Line Chart: To track Netflix content trends over the years.
    - Bar Charts: To identify the Top 10 Countries and the most common Content Ratings.
    - Heatmap: To show how investment in different genres changed over time.
    - Density Plot: To visualize the "age" of content on the platform.

7. Final Results & Insights

    Our analysis led to three major conclusions:

    - Movie Dominance: Movies make up nearly 70% of the library, but additions have been slowing down since 2019.
    - Mature Content Focus: The library is heavily weighted toward TV-MA (Mature) audiences, which might limit its appeal to families.
    - The Strategic Risk: Netflix is facing a "one-and-done" problem. Because they have so many movies and mature dramas compared to long-running TV shows, they risk higher churn rates as users finish a single hit and then cancel their subscription.
